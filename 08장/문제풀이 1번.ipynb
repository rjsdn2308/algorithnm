파이썬 sort() = Timsort 알고리즘
Timsort란?
Merge Sort + Insertion Sort를 혼합한 하이브리드 정렬 알고리즘
실제 데이터의 특징을 분석하여 최적의 방식으로 동작
Python의 list.sort()와 sorted() 모두 Timsort 사용

왜 Timsort가 빠르고 좋은가?
1. 이미 정렬된 부분을 감지(Run Detection)
데이터에서 이미 정렬된 구간(run)을 자동으로 찾아서 이용함 → 매우 빠름
실제 데이터는 부분적으로 정렬된 경우가 많기 때문에 효율적.

2. 최악의 경우에도 안정적인 O(N log N)
Merge sort 기반이므로 최악의 경우 성능 보장.

3. 거의 정렬된 데이터에 매우 강함
거의 정렬돼 있으면 Insert sort처럼 O(N) 성능을 낼 수 있음.

4. Stable Sort(안정 정렬)

같은 값의 원소 순서가 유지됨.

📌 복잡도
경우	시간 복잡도
최선	O(N)
평균	O(N log N)
최악	O(N log N)
공간 복잡도	O(N)
